# Search

```python
from searchUtils import findExt, findPattern, findPatternExt, findDirs, findDirsPattern
```

```python
files = findExt(basedir=basedir, ext="*.p")
print(files == ['/Users/tgadfort/Dropbox/ACTORS.p', '/Users/tgadfort/Dropbox/ACTRESSES.p'])
files = findPattern(basedir=basedir, pattern="ACTOR")
print(files == ['/Users/tgadfort/Dropbox/ACTORS.p'])
files = findPatternExt(basedir, pattern="ACTR", ext="*.p")
print(files == ['/Users/tgadfort/Dropbox/ACTRESSES.p'])

idir = setDir(basedir, "test")
dirs = findDirs(idir)
print(dirs == ['/Users/tgadfort/Dropbox/test/nslook'])
dirs = findDirsPattern(basedir=basedir, pattern="test")
print(dirs == ['/Users/tgadfort/Dropbox/test',
 '/Users/tgadfort/Dropbox/Latest',
 '/Users/tgadfort/Dropbox/GreatestShowmanSoundtrack'])
```
